# Feature Engineering

## Null Value Resolution

Data regarding high school financials or other statistics is frequently omitted in the case of international and private high schools due to lack of information accessibility or relevance. In these specific cases, the original authors of the dataset use the phrase "Data not available" to signal this in conjunction with `np.NaN` for other missing values.   

The presence of this string prevents casting columns to the float datatype, so it must be reconciled before feature engineering and imputation. "Data not available" will be globally replaced with `np.NaN` and treated as such. These values will be imputed like all other missing values in the modeling pipeline.

```python
# Replace "Data not available" with np.Nan
df = df.replace({"Data not available" : np.NaN})
df.isnull().sum()
```

```python
df.dtypes
```

```python
# Convert 'Tuition of High School' to float
df['Tuition of High School'] = df['Tuition of High School'].str.replace(",","").astype(float)
```

```python
# Convert 'Tuition of High School' to float
df['Free and reduced percentage of  High School'] = df['Free and reduced percentage of  High School'].astype(float)
```

```python
df.dtypes
```

```python
# Remap 'Greek Council' to binary classification
df['greek'] = df['Greek Council']
df['greek'].value_counts()
```

```python
# Consolidate Sorority/Fraternity and map to binary classification
df['greek'] = df['greek'].map({'None':'No', 'Sorority':'Yes', 'Fraternity':'Yes'})
df['greek'].value_counts()
```

```python
# Fix errors in 'Sports Team', Reengineer to universal
df['Sports Team'].value_counts()
```

```python
df['Sports Team'] = df['Sports Team'].str.replace('Fooball', 'Football')
```

```python
# Consolidate Men and Women's sports
df['sports_consolidated'] = df['Sports Team']
df['sports_consolidated'] = df['sports_consolidated'].str.lstrip("Men's ")
df['sports_consolidated'] = df['sports_consolidated'].str.lstrip("Women's ")
# df['sports_consolidated'].value_counts()
```

```python
# Fix 'restling' misspelling
df['sports_consolidated'] = df['sports_consolidated'].str.replace('restling', 'Wrestling')
df['sports_consolidated'].value_counts()
```

# Modeling

## Create Baseline Model

```python
from sklearn.model_selection import train_test_split
# Train / Test Split
train, test = train_test_split(df, test_size=0.20, random_state=20)
train.shape, test.shape
```

```python
# Train / Val Split
train, val = train_test_split(train, random_state=20)
train.shape, val.shape
```

```python
# Majority Classification Baseline
train['greek'].value_counts(normalize=True)
```

```python
cols = train.columns.tolist()
cols
```

Column Exclusion Justifications:  
**Greek Council**: Re-engineered as 'Greek' in binary classification  
**Greek Organization**: Would introduce data leakage  
**Home City**: Redundant with Latitude and Longitude  
**Home State**: Redundant with Latitude and Longitude  
**Home Country**: Redundant with Latitude and Longitude  
**Sports Team**: Consolidated into unisex categories

Unfortunately, Organizational data can't be used for modeling purposes. Including this information would directly inform the model of involvement, introducing data leakage. Due to lack of more specific information about both the students and organizations, there is not enough data to provide classifications for individual organizations. 

```python
# Create feature matrices and target vectors

drop = ['Greek Council', 'Greek Organization', 'greek', 
        'Home City', 'Home State', 'Home Country']
target = 'greek'

X_train = train.drop(columns=drop)
y_train = train[target]

X_val = val.drop(columns=drop)
y_val = val[target]

X_test = test.drop(columns=drop)
y_test = test[target]
```

# Random Forest Model

```python
from sklearn.model_selection import train_test_split
import category_encoders as ce
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
import category_encoders as ce
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
# Create tree based model pipeline
tree_pipeline = make_pipeline(
    ce.OrdinalEncoder(),
    SimpleImputer(strategy='median'),
    RandomForestClassifier(random_state=20, n_jobs=-1)
)

# Create hyperperameter dictionary for RandomizedSearchCV()
param_distributions = {
    'simpleimputer__strategy': ['mean', 'median'],
    'randomforestclassifier__n_estimators': randint(50, 500),
    'randomforestclassifier__max_depth': [15, 20, 25, 30, None],
}

search = RandomizedSearchCV(
    tree_pipeline,
    param_distributions=param_distributions,
    n_iter=20,
    cv=5,
    scoring='accuracy',
    verbose=0,
    return_train_score=True,
    n_jobs=-1,
    random_state=20
)

search.fit(X_train, y_train);
# tree_pipeline.fit(X_train, y_train)
```

```python
print('Best hyperparameters', search.best_params_)
print('Cross-validation Accuracy', search.best_score_)
```

```python
# # Results of CV
# pd.DataFrame(search.cv_results_).sort_values(by='rank_test_score').T
```

```python
# Save results of hyperperameter optimization to var
tree_pipeline = search.best_estimator_
```

## Random Forest Metrics

```python
print('Train Accuracy', tree_pipeline.score(X_train, y_train))
# Validation accuracy_score
print('Validation Accuracy', tree_pipeline.score(X_val, y_val))
```

```python
# Validation ROC/AUC
from sklearn.metrics import roc_auc_score
y_pred_proba = tree_pipeline.predict_proba(X_val)[:,-1]
val_roc = roc_auc_score(y_val, y_pred_proba)

print('Validation ROC/AUC: ', val_roc)
```

## Random Forest Model Visualizations

```python
# Plot confusion matrix
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(
    tree_pipeline, 
    X_val, 
    y_val, 
    values_format='.0f', 
    xticks_rotation='vertical'
)
plt.show()
```

```python
# Permutation importances
import eli5
from eli5.sklearn import PermutationImportance

# Instantiate
permuter = PermutationImportance(
    model,
    scoring='accuracy',
    n_iter=5,
    random_state=20
)

permuter.fit(X_val_transformed, y_val)
```

```python
# Set column names to list
feature_names = X_val.columns.tolist()

#Print Permutation Importances
eli5.show_weights(
    permuter,
    top=None,
    feature_names=feature_names
)
```
