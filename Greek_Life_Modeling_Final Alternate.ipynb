# Feature Engineering

**Null Value Resolution**  
Data regarding high school financials or other statistics is frequently omitted in the case of international and private high schools due to lack of information accessibility or relevance. In these specific cases, the original authors of the dataset use the phrase "Data not available" to signal this in conjunction with `np.NaN` for other missing values.   

The presence of this string prevents casting columns to the float datatype, so it must be reconciled before feature engineering and imputation. "Data not available" will be globally replaced with `np.NaN` and treated as such. These values will be imputed like all other missing values in the modeling pipeline.

```python
def feature_engineering(X):
    
    # Avoid warnings from changes to original data structure
    df = X.copy()
    
    # Replace "Data not available" with np.Nan
    df = df.replace({"Data not available" : np.NaN})
        
    # Convert 'tuition' to float
    df['tuition'] = df['tuition'].str.replace(",","").astype(float)
        
    # Convert 'free_lunch_percent' to float
    df['free_lunch_percent'] = df['free_lunch_percent'].astype(float)
        
    # Consolidate Sorority/Fraternity and map to binary classification
    df['greek'] = df['greek_council'].map({'None':'No', 'Sorority':'Yes', 'Fraternity':'Yes'})
    
    # Fix errors in 'Sports Team', Reengineer to genderless
    df['sports_team'] = df['sports_team'].str.replace('Fooball', 'Football')
    df['sports'] = df['sports_team']
    df['sports'] = df['sports'].str.lstrip("Men's ")
    df['sports'] = df['sports'].str.lstrip("Women's ")
    df['sports'] = df['sports'].str.replace('restling', 'Wrestling')
    
    return df
```

```python
# Execute feature engineering function
df = feature_engineering(df)

# Ensure proper dtypes were cast
df.dtypes
```

# Modeling

## Data Subsetting

Create train, validation, and test subsets of data for modeling by using train_test_split() twice

```python
from sklearn.model_selection import train_test_split
```

```python
# Split df into train and test sets
train, test = train_test_split(df, test_size=0.20, random_state=20)
train.shape, test.shape
```

```python
# Split train into train and validation sets
train, val = train_test_split(train, random_state=20)
train.shape, val.shape
```

**Feature Exclusions**:  
`greek`: Classification target  
`greek_council`: Re-engineered to binary classification, `greek` is used instead as target   
`greek_organization`: Would introduce data leakage  
`city`: Redundant with latitude and longitude  
`state`: Redundant with latitude and longitude  
`country`: Redundant with latitude and longitude  

**Data Redundancy**  
Not only do the features `city`, `state`, and `country` contain the same information as `latitude` and `longitude`, but they also hold the information in a format that is not numerical or in a geographically consistent system. All city, state, and country location information is inherently held in the corresponding latitude and longitude coordinates. Attempting to encode these columns would only result in a worse latitude/longitude coordinate system that is not geographically consistent.

**Data Leakage**  
The data in the `greek_organization` feature can not be used for modeling purposes. The values in this column are specific subdivisons of the original values in `greek_council`. Including this information would directly inform the model of the student classification, introducing data leakage.

```python
# Organize data into feature matrices and target vectors

# Feature matrix exclusions
drop = ['greek', 'greek_council', 'greek_organization', 
        'city', 'state', 'country']

# Target
target = 'greek'

# Train set
X_train = train.drop(columns=drop)
y_train = train[target]

# Validation set
X_val = val.drop(columns=drop)
y_val = val[target]

# Test set
X_test = test.drop(columns=drop)
y_test = test[target]
```

## Random Forest Model

```python
import category_encoders as ce
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from scipy.stats import randint
```

**Model in pipeline**

```python
# Create tree based model sklearn pipeline
tree_pipeline = make_pipeline(
    # Encode objects/categoricals
    ce.OrdinalEncoder(),
    # Impute null values
    SimpleImputer(strategy='median'),
    # Model using RandomForest
    RandomForestClassifier(random_state=20, n_jobs=-1)
)

# Create hyperparameter dictionary for RandomizedSearchCV()
param_distributions = {
    'simpleimputer__strategy': ['mean', 'median'],
    'randomforestclassifier__n_estimators': randint(50, 500),
    'randomforestclassifier__max_depth': [15, 20, 25, 30, None],
}

# Use RandomizedSearchCV() to optimize hyperparameters
# Instantiate the search object
search = RandomizedSearchCV(
    tree_pipeline,
    param_distributions=param_distributions,
    n_iter=20,
    cv=5,
    scoring='accuracy',
    verbose=0,
    return_train_score=True,
    n_jobs=-1,
    random_state=20
)

# Perform tuning
search.fit(X_train, y_train);
```

```python
# Get optimized parameters
print('Best hyperparameters', search.best_params_)
print('Cross-validation Accuracy', search.best_score_)
```

```python
# Save optimized pipeline
tree_pipeline = search.best_estimator_
```

**Model outside of pipeline**

Many visualization libraries do not yet support scikit-learn's pipeline, requiring the model be reconstructed outside of the pipeline. 

```python
# Instantiate encoder and fit
encoder = ce.OrdinalEncoder()
X_train_encoded = encoder.fit_transform(X_train)
X_val_encoded = encoder.transform(X_val)

# Instantiate imputer and fit
imputer = SimpleImputer(strategy='median')
X_train_imputed = imputer.fit_transform(X_train_encoded)
X_val_imputed = imputer.transform(X_val_encoded)


# Instantiate Model using params from tuning
model = tree_pipeline['randomforestclassifier']

# Model Fit
model.fit(X_train_imputed, y_train)
```

## Random Forest Metrics

Although not used to evaluate the model's performance, train accuracy is included to function as a rudimentary overfitting signal. 

```python
# Train accuracy score
print('Train Accuracy:', tree_pipeline.score(X_train, y_train))
```

```python
from sklearn.metrics import roc_auc_score
```

```python
# Create function to return validation metrics
def val_metrics(pipeline, X_val, y_val):
    """Calculates accuracy and roc auc scores, then prints"""
    
    # Get validation accuracy
    val_accuracy = pipeline.score(X_val, y_val)
    # Get list of predicted probabilities
    y_pred_proba = pipeline.predict_proba(X_val)[:,-1]
    # Calculate ROC AUC Score
    val_roc = roc_auc_score(y_val, y_pred_proba)
    
    # Print
    print('Validation Accuracy Score: ', val_accuracy)
    print('Validation ROC AUC: ', val_roc) 
```

```python
# Execute val_metrics to get accuracy and ROC AUC for validation set
val_metrics(tree_pipeline, X_val, y_val)
```

### 1 Feature PDP - Tuition

```python
# PDP can't handle NaN values, so X_val must be encoded then imputed
# PDP also requires a DataFrame, so the imputed X_val array is recast to a DataFrame
X_val_encoded_df = pd.DataFrame(data=X_val_imputed, columns=X_val.columns)
```

```python
# 1 feature partial dependence plot
from pdpbox.pdp import pdp_isolate, pdp_plot
from pdpbox import pdp

# Set feature to analyze partial dependence
feature = 'tuition'

# Create isolated object with attributes
pdp_dist = pdp_isolate(
    # Regressor used
    model=model,
    # Current trend is using validate set here
    dataset=X_val_encoded_df,
    # List of other features to freeze
    model_features=X_val.columns,
    # Analyzed feature from abovev
    feature=feature
)
```

```python
# Create Partial Dependency Plot
pdp_plot(pdp_dist, feature);
plt.xlim(0,40000)
plt.show()
```

This Partial Dependency Plot for `tuition` shows that as the tuition of high school increases, the RandomForestClassifier model increases its predicition probability for the student to be involved in Greek Life.  

Although the graph begins to decline near the 40,000 dollar tuition threshold, this is not relevant because there are only a handful of students who make up the population with tuition that high, and the model is not trained on that range.

# Test Set Metrics

```python
from sklearn.metrics import accuracy_score
```

```python
def test_metrics(model_name, pipeline, X_test, y_test):
    """Calculates test accuracy, then prints.
    model_name is a string."""
    
    # Calculate y_pred for test set
    y_pred = pipeline.predict(X_test)
    # Calculate accuracy
    acc = accuracy_score(y_test, y_pred)
    # Print
    print(model_name +' model test accuracy: ', acc)
```

**RandomForestClassifier**

```python
test_metrics("RandomForestClassifier", tree_pipeline, X_test, y_test)
```

**LogisticRegression**

```python
test_metrics("LogisticRegression", log_pipeline, X_test, y_test)
```

Both the linear and tree based model have validation and test accuracies higher than that of the majority class baseline. The linear model averages about 3% more accurate; however, it is in some ways held back by it's inflexibility with encoding schemes that may prove better suited for the tree based model if the data's horizon's were broader.
